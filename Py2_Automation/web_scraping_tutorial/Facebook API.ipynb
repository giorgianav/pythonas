## Get the list of friends

```python
import facebook
import random

access_token = '...'
graph = facebook.GraphAPI(access_token, version='2.9')


# Get the list of my friends
myFriends = graph.get_object("me/friends") 

friend_id = None
your_friend_name = '...'

while  True:

    # The 'data' field contains a list with the returned friends
    for f in myFriends['data']:

        # Check if the name is the one we are searching for
        if f['name'] == your_friend_name:

            # Keep the our friend's id to be able to send other requests
            friend_id = f['id']
            break

    # Stop when we reach the end of list of friends
    try:
        myFriends['paging']['next']
    except KeyError as e:
        print ("No more friends")
        break

    # Go to the next page of friends
    next_page = myFriends['paging']['cursors']['after']

    # Get the friends from the next page
    myFriends = graph.get_object("me/friends?after=" + next_page)

# Get your friend albums
albums = graph.get_object(friend_id + "/albums")
album_id = None
album_name = '...'

# Find a specific album 
for a in albums['data']:
    if a['name'] == album_name:
        album_id = a['id']
        break

        
print ("Album ID:")
print (album_id)

# Get the photos from that album
photos = graph.get_object(album_id + "/photos")

# A list of possible comments
comments = ['Yeah, summer is here!',
            'Ce poză faină ai!',
            'Enjoy every moment.',
            'Only great minds can afford a simple style.',
            'The greatest power is often simple patience.',
            'No rain, no flowers.',
            'Life is only as good as your mindset.',
            'Be kind!',
            'Keep it simple.',
            'That moment was amazing!',
            'Love your style!',
            'Great experience!',
            'Be happy! Be Olaf!'
            'Whaaaaat?!',
            'Hahahaa!']

# Leave a comment on each photo from that album
for p in photos['data']:
    c = random.choice(comments)
    photo_id = p['id']
    print (photo_id)
#     graph.put_comment(object_id=photo_id, message=c)
    
```
